```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Bivariate case 

mean = np.array([0.0,0.0])
cov = np.array([[1,0],
                [0,1]])   # here we have defined 0 cov x and y are independent of each other as shown in below scatter plot 
                          # which is like correlation b/w color and height of monkey ie 0 correlation                       
    
dist = np.random.multivariate_normal(mean,cov,500)
```

```python
print(dist.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1]) 
plt.show()  # the distribution is centered around mean coz of o correlation
```

```python
# defining a positive corrilation ex height and weight are directly proportional to each other

mean = np.array([0.0,0.0])
cov = np.array([[1,0.5],
                [0.5,1]])                       
    
dist = np.random.multivariate_normal(mean,cov,500)

plt.scatter(dist[:,0],dist[:,1]) 
plt.show()  # it is slightly tilted towards line x=y coz of +ve distribution
```

```python
# more +ve corelation results in more tilted scattering

mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],
                [0.8,1]])                       
    
dist = np.random.multivariate_normal(mean,cov,500)

plt.scatter(dist[:,0],dist[:,1]) 
plt.show()
```

```python
# visualise two distribution simultaneously

mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],
                [0.8,1]])                       

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],
                [0.2,1.1]])   


dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)

plt.scatter(dist[:,0],dist[:,1]) 
plt.scatter(dist2[:,0],dist2[:,1]) 

plt.show()
```

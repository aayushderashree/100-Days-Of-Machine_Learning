```python
xsn = np.random.randn(100)

sigma = 15
u = 60

x = xsn*sigma + u

```

```python
print(xsn) # most of values are in range -1 to 1
```

```python
 print(x) # most values will be in range 55 to 65 
```

```python
 # to conver into interger 

x = np.round(xsn*sigma+u)

print(x)

x2 = np.round(xsn*5+40)

print(x2)
```

Histogram

```python
plt.style.use("seaborn")
plt.hist(x,alpha = 0.9,label = "physics")
plt.hist(x2,alpha=0.8, label = "maths")
plt.xlabel("marks range")
plt.ylabel("prob/freq")
plt.title("histogram")
plt.legend()
plt.show()
```

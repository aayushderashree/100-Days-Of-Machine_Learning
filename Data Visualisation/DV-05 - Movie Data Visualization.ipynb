```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# READ A CSV FILE 

df =  pd.read_csv("movie_metadata.csv") 
```

```python
titles = df.get("movie_title")

print(titles)

print(type(titles))
```

```python
# typecast obj.series into a python list

titles = list(df.get('movie_title'))
```

```python
print(titles[:5])
```

```python
# remove the last 3 characters

print(titles[0][:-1])
```

```python
# Making a dictionary

freq_title = {}

for title in titles:
    length = len(title)
    
    if freq_title.get(length) is None:
        freq_title[length] = 1
    
    else:
        freq_title[length] += 1 
        
```

```python
print(freq_title)
```

```python
# now we'll make a graph

x = np.array(list(freq_title.keys()))

y  = np.array(list(freq_title.values()))
```

```python
print(x,y)
```

```python
plt.style.use("seaborn")

plt.scatter(x,y)
plt.xlabel("length of movie title")
plt.ylabel("freq count of movies")
plt.title("movie data visualisation problem")

plt.show()
```

# Bar Graphs

```python
import matplotlib.pyplot as plt 

import numpy as np
```

```python
plt.bar?
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25 ,[10,20,15], width = 0.5, label = "current year", tick_label = ["gold","platinum","silver"]) # prices current year
plt.bar(x_coordinates+0.25 ,[20,10,12], width = 0.5, label = "next year") # prices next year

plt.title("metal plot comparision")
plt.xlabel("metal")
plt.ylabel("price")
plt.legend()
 
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25 ,[10,20,15], width = 0.5, label = "current year", tick_label = ["gold","platinum","silver"],color =  "green") # prices current year
plt.bar(x_coordinates+0.25 ,[20,10,12], width = 0.5, label = "next year",color = "red") # prices next year

plt.title("metal plot comparision")
plt.xlabel("metal")
plt.ylabel("price")
plt.ylim(0,40) # to set range in graph and values will get set accordingly
plt.legend()
 
plt.show() 

```

```python
subjects = "maths", "english", "physics", "chem"
weightage = [10,20,15,5]

plt.pie(weightage, labels = subjects, explode = [0,0,0.1,0], autopct = '%1.1f%%', shadow = True)
plt.title = "subjects"

plt.show()
```

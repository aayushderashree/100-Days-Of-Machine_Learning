```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
themes = plt.style.available

print(themes)
```

```python
plt.style.use("seaborn")
```

```python
 # line plot 
    
x = np.arange(10)

y1 = x**2

y2 = 2*x+3
 
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y2,color='red', label = "apple", marker="o")

plt.plot(x,y1,color='green', label = "kiwi", linestyle="dashed", marker = "*")

plt.xlabel("time")

plt.ylabel("price")

plt.title("prices of fruits over time")

plt.legend()

plt.show()

# without calling plot.show() we can produce output using jupyter but in general python only an object will be shown without it
# also every line we use plot.show() we are gonna get a new graph
```

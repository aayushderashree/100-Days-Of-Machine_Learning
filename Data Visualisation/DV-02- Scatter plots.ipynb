# Scatter plots

instead of line plot we wand discreate points

```python
x = np.arange(10)

y1 = x**2

y2 = 2*x+3
```

This is the sample equations from line plots

plt.plot(x,y2,color='red', label = "apple", marker="o")

plt.plot(x,y1,color='green', label = "kiwi", linestyle="dashed", marker = "*")

plt.xlabel("time")

plt.ylabel("price")

plt.title("prices of fruits over time")

plt.legend()

plt.show()

```python
# adjust size of plot

plt.figure(figsize = (5,10))

plt.scatter(x,y2,color='red', label = "apple", marker="o")

plt.scatter(x,y1,color='green', label = "kiwi", linestyle="dashed", marker = "*")

plt.xlabel("time")

plt.ylabel("price")

plt.title("Scatter plot - prices of fruits over time")

plt.legend()

plt.show()
```
